Research Question # 1: Is there an association between how often you have trouble sleeping and if you are happy in your role as a parent?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
````

Codebook 1: Trouble_Sleeping

Trouble Sleeping Codebook.PNG

````python
#Manage data for variable 1 (use multiple lines)
#Variable 1: Trouble_Sleeping
df['H4SP6'].value_counts()
````

````python
#Create variable 1
df['Trouble_Sleeping'] = df['H4SP6']
````

````python
#Remove response of not on test, refused and I don't know (5, 6 and 8) and remove missing values
df['Trouble_Sleeping'] = df['Trouble_Sleeping'].loc[ df['Trouble_Sleeping'] !=' ']
df['Trouble_Sleeping'] = df['Trouble_Sleeping'].loc[ df['Trouble_Sleeping'] !='5']
df['Trouble_Sleeping'] = df['Trouble_Sleeping'].loc[ df['Trouble_Sleeping'] !='6']
df['Trouble_Sleeping'] = df['Trouble_Sleeping'].loc[ df['Trouble_Sleeping'] !='8']
df['Trouble_Sleeping'].value_counts()
````

````python
#Format names from numbers
df['Trouble_Sleeping'].replace(['0','1','2','3','4'], ['Never','Less than once a week','Once or twice a week','Two or three times a week','Five or more'] , inplace = True)
df['Trouble_Sleeping'].value_counts()
````

````python
#Graphs and tables or statistics
````

````python
#Pie Chart: Add titles
plt.title("How often do you have trouble staying asleep at night?")
df['Trouble_Sleeping'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown', 'palegoldenrod'])
````

````python
#Bar Graph: Add Axis titles
plt.xlabel("Trouble_Sleeping")
plt.ylabel("Percent")
plt.title("How often do you have trouble staying asleep at night?")
df['Trouble_Sleeping'].value_counts(normalize = True).plot(kind ='bar', color = 'maroon', width = 0.2, alpha = 0.5)
````

Codebook 2: Happy_Parent

Happy Parent Codebook.PNG

````python
#Manage data for variable 2: Happy_Parent
df['H4KK15A'].value_counts()
````

````python
#Create new variable 'Happy_Parent'
df['Happy_Parent'] = df['H4KK15A']
df['Happy_Parent'].value_counts()
````

````python
#Remove response of not on test, refused and I don't know (96, 97 and 98) and remove missing values
df['Happy_Parent'] = df['Happy_Parent'].loc[ df['Happy_Parent'] !=' ']
df['Happy_Parent'] = df['Happy_Parent'].loc[ df['Happy_Parent'] !='96']
df['Happy_Parent'] = df['Happy_Parent'].loc[ df['Happy_Parent'] !='97']
df['Happy_Parent'] = df['Happy_Parent'].loc[ df['Happy_Parent'] !='98']
df['Happy_Parent'].value_counts()
````

````python
#Format names from numbers
df['Happy_Parent'].replace(['1','2','3','4','5'], ['Strongly Agree','Agree','Neither Agree nor Disagree','Disagree','Strongly Disagree'] , inplace = True)
df['Happy_Parent'].value_counts()
````

````python
#Graphs and tables or statistics
````

````python
#Pie Chart: Add titles
plt.title("Are you happy in your role as a parent?")
df['Happy_Parent'].value_counts().plot(kind ='pie', autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown', 'palegoldenrod'])
````

````python
#Bar Graph: Add titles
plt.title("Are you happy in your role as a parent?")
df['Happy_Parent'].value_counts(normalize = True).plot(kind ='bar', color = 'rosybrown', width = 0.2)
````
